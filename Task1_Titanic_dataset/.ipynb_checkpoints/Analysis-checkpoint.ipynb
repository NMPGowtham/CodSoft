```python
updated_features = ["Survived","Pclass","Age","SibSp","Parch","Fare","Embarked","Sex"]
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(9, 6))
axes = axes.flatten()
for ax, feature in zip(axes, updated_features):
    sns.histplot(df[feature], kde=True, ax=ax)
    ax.set_title(feature)
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_val = df[column].median()
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), median_val, df[column])
    return df
```

```python
updated_df = remove_outliers(df,"Age")
```

```python
updated_df.describe()
```

```python
updated_df["Fare"] = np.log1p(updated_df["Fare"])
updated_df.describe()
```

```python
def Gender2bin(data):
    if data == "male":
        return 0
    if data == "female":
        return 1
    
```

```python
def City2bin(data):
    if data == "C":
        return 0
    if data == "Q":
        return 1
    if data == "S":
        return 2
    if data == "U":
        return 3
```

```python
temp = updated_df.copy()
temp["Embarked"] = temp["Embarked"].apply(City2bin)
updated_df = temp
updated_df
```

```python
temp = updated_df.copy()
temp["Sex"] = temp["Sex"].apply(Gender2bin)
updated_df = temp
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(9, 6))
axes = axes.flatten()

for ax, feature in zip(axes, updated_features):
    mean_val = updated_df[feature].mean()
    sns.histplot(updated_df[feature], kde=True, ax=ax)
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
    ax.set_title(feature)

plt.tight_layout()
plt.show()
```
